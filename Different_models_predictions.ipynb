```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats
```

```python
import matplotlib 
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/DriveMyScream/Python_libraries/main/weatherAUS.csv")
```

```python
df
```

```python
df['RainTomorrow'].fillna(df['RainTomorrow'].mode()[0], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['RainTomorrow', 'Date'], axis=1), df['RainTomorrow'], random_state=42, test_size=0.25)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
categorical_cols = X_train.select_dtypes('object').columns.tolist()
categorical_cols
```

```python
X_train[categorical_cols].nunique()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

# Handling numerical values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
X_train[numeric_cols].isna().sum()
```

```python
imputer.fit(X_train[numeric_cols])
```

```python
X_train[numeric_cols] = imputer.transform(X_train[numeric_cols])
X_test[numeric_cols] = imputer.transform(X_test[numeric_cols])
```

```python
X_train[numeric_cols].isna().sum()
```

# Handling categorical value

```python
imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
X_train[categorical_cols].isnull().sum()
```

```python
imputer.fit(X_train[categorical_cols])
```

```python
X_train[categorical_cols] = imputer.transform(X_train[categorical_cols])
X_test[categorical_cols] = imputer.transform(X_test[categorical_cols])
```

```python
X_train[categorical_cols].isnull().sum()
```

# Encoding variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
le.classes_
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X_train[categorical_cols].nunique()
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
encoder.fit(X_train[categorical_cols])
```

```python
encoder.categories_
```

```python
encoded_cols = list(encoder.get_feature_names(categorical_cols))
encoded_cols
```

```python
X_train[encoded_cols] = encoder.transform(X_train[categorical_cols])
X_test[encoded_cols] = encoder.transform(X_test[categorical_cols])
```

```python
pd.set_option('display.max_columns', None)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train[numeric_cols])
```

```python
X_train[numeric_cols] = scaler.transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
X_train
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train[numeric_cols + encoded_cols], y_train)
```

```python
y_pred = model.predict(X_test[numeric_cols + encoded_cols])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print("Logistic Regression",np.mean(cross_val_score(model, X_train[numeric_cols + encoded_cols], y_train, scoring='accuracy',cv=10)))
```
